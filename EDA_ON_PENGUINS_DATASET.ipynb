**VISUALIZATION:**

**COUNTPLOT**

```python
plt.title('No. of Species')
plot=sns.countplot(x="species",data=data)
```

From this visualization,we can observe that our dataset is not balanced as we don't have equal number of count of the species. We can also observe that we have more number of Adelie species of the penguins maintaining a huge differnce with the count of the Chinstrap.

**SCATTER PLOT**

```python
plot=sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',hue="species",data=data)
```

From the above visualization, we can tell that the Adelie species has smaller bill length but higher depth. While we see Chinstrap species lies in almost middle for length as well as depth. And the Gentoo species has larger bill lengths and smaller bill depths.

**STRIP PLOT**

```python
fig, axs = plt.subplots(figsize = (6,6))
plot=sns.stripplot(x="species",y="bill_length_mm",hue="sex",data=data);
```

From the above visualization,we observed that the male penguins has higher bill length when compared to female penguins in all the three species of our dataset.

**BOX PLOT**

```python
fig, axs = plt.subplots(figsize = (8,6))
plot=sns.boxplot(x="species",y="body_mass_g",data=data)
```

Here,we can observe that both Adelie and Chinstrap species have almost equal range of body masses.So,we can only differentiate the Gentoo species from the above box plot.We can also observe the outliers present among the Chinstrap species.

**Correlation between variables**

```python
data.corr()
```

Further, we can visualise this correlation using a heatmap.

**HEAT MAP**

```python
fig = plt.figure(figsize = (15,9))
sns.heatmap(data.corr(), cmap='Reds', annot = True);
```

From the above heatmap, we see that flipper_length and body mass have a high correlation, body mass and bill length have good correlation as well as flipper_length and bill_length have good correlations.

**HIST PLOT**

```python

data.hist(edgecolor='black',bins = 25, figsize= (12,6))
plt.show()
```

Here,we can visualize the distribution of length,depth and mass of our dataset.

**PAIR PLOT**

```python
sns.pairplot(data = data,hue="species", height=3)
```

pairplot is showing the bivariate relation between each pair of features. From the pairplot, we'll see that the Gento species is separataed from the other two across all feature combinations

**VIOLIN PLOT**

```python

sns.violinplot(x="species", y="body_mass_g", data=data, size=6)
plt.show()
```

Here,we can observe that both Adelie and Chinstrap species have almost equal range of body masses. So,we can only differentiate the Gentoo species from the above violin plot. We can also observe the outliers present among the Chinstrap species

**CONCLUSION**


*   We have four numerical columns while three categorical columns
*   The dataset is not balanced i.e. equal records are not present for all three species.
*   A strong correlation is present between flipper length and body mass.
*   The Adelie and Chinstrap species are usually mixed and are sometimes hard to separate.
*   The Gentoo species is the most easily distinguishable because of its large body masses.
